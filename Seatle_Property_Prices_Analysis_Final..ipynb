# Multiple Linear Regression 

 Thi is an extension of linear regression (OLS) that uses just one explanatory variable(used above).

```python
x = ols_data[['const', 'sqft_living', 'bathrooms', 'grade', 'sqft_above', 'sqft_living15']]
y = ols_data['price']
```

```python
model = sm.OLS(y, x).fit()
```

```python
model.summary()
```

```python
data.groupby('zipcode')['price'].median().reset_index().sort_values('price', ascending = False)
```

We will now test that algorithm by the most expensive zipcode.

```python
expensive = ols_data.loc[data['zipcode'] == 98039]
```

```python
expensive
```

```python
x = expensive[['sqft_living']]
y = expensive[['price']]
```

```python
regressor = LinearRegression()
regressor.fit(x, y)
```

```python
plt.figure(figsize=(15,8))
plt.ticklabel_format(axis='y', style='plain')
plt.scatter(x, y)
plt.plot(x, regressor.predict(x))
```

```python
regressor.score(x, y)
```

Accurancy score increases to 88% from 54%.

```python
np.array(range(20,25))
```

```python
expensive.corr()
```

Can we make it much better with more independent variables against price?

```python
x = expensive[['const', 'sqft_living', 'bathrooms', 'grade', 'yr_built', 'sqft_lot', 'sqft_lot15']]
y = expensive['price']
```

```python
model1 = sm.OLS(y, x).fit()
```

```python
model1.summary()
```

To best predict price determinants for properties in this dataset, Multi-Linear Regression algrorithm suits best. It predicts that 92% of the dependent variable is determined by more than 6 independent variables such as sqft_living, bathrooms, grade, yr_built, sqft_lot, sqft_lot15 among others.
